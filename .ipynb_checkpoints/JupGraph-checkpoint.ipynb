```python
#увеличим дефолтный размер графиков
from pylab import rcParams
rcParams['figure.figsize'] = 15,12

# импортируем нужные библиотеки
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

Функция, которая загружает указанные года

```python
def load_years(years_list):
    names_by_year = {}
    for year in years_list:
        names_by_year[year] = pd.read_csv(
            f'names/yob{year}.txt',
            names=['Name','Gender','Count']
        )
    names_all = pd.concat(names_by_year).reset_index(level=1, drop=True)
    names_all = names_all.rename_axis(index='Year', columns="")
    return names_all
    
```

__1 Построить график изменения количества имен Ruth и Robert с 1900 по 2000.__

```python

df = load_years(list(range(1900, 2000)))
res = df[(df['Name'] == 'Ruth') | (df['Name'] == 'Robert')].groupby(['Year','Name']).sum().sort_values('Year', ascending=True)
```

строим график

```python
res.unstack('Name').plot()
```

__Построить ~~гистограмму~~ столбчатую диаграмму по количеству их имен с 1900 по 2000 с 5-летними промежутками (1900, 1905, 1910, …, 1995, 2000).__

```python
res[0::10].unstack("Name").plot(kind='bar')
```

что-то не очень, попробуем немного усложнить

__Построить круговую диаграмму по количеству употреблений для ТОП-10 популярных имен, начинающихся на R, за 1950 год.__

```python
df = load_years([1950])
res = df[df.Name.str.contains('R')].groupby(['Name']).sum().sort_values('Count', ascending=False).head(10)
res.plot(kind='pie', y='Count')
```
